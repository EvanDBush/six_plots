```python
import pandas as pd 
df = pd.read_csv("data/coffee_sales_1.csv")
df
```

```python
#drops date column. information preserved in datetime column
df = df.drop(columns=['date'])

#converts datetime column from object to datetime object.
df['datetime'] = pd.to_datetime(df['datetime'])

#transforms money column into more familiar number for US readers
df['money'] = (df['money']/10).round(2)

```

```python
#writes dataframe to csv file
path = 'data/clean_coffee_data.csv'
df.to_csv(path, index=False)
```
